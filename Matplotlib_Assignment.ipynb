## Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
##### plot a scatter plot.
##### import numpy as np
##### np.random.seed(3)
##### x = 3 + np.random.normal(0, 2, 50)
##### y = 3 + np.random.normal(0, 2, len(x))
##### Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel('X_Axis')
plt.ylabel('Y_Axis')
plt.title('Random_Data')
```

## Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
#### import numpy as np
#### For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
#### For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
#### For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
#### For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

#### Answer - We use subplot() function for draw multiply plots.

```python
plt.subplot(2,2,1)
plt.plot(np.array([0, 1, 2, 3, 4, 5]),np.array([0, 100, 200, 300, 400, 500]))
plt.title('First_plot')
plt.show

plt.subplot(2,2,2)
plt.plot(np.array([0, 1, 2, 3, 4, 5]),np.array([50, 20, 40, 20, 60, 70]))
plt.title('Second_plot')
plt.show

plt.subplot(2,2,3)
plt.plot(np.array([0, 1, 2, 3, 4, 5])  ,np.array([10, 20, 30, 40, 50, 60]))
plt.title('Third_plot')
plt.show

plt.subplot(2,2,4)
plt.plot(np.array([0, 1, 2, 3, 4, 5]) , np.array([200, 350, 250, 550, 450, 150]))
plt.title('Fouth_plot')
plt.show
```

## Q4What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
#### import numpy as np
#### company = np.array(["Apple", "Microsoft", "Google", "AMD"])
#### profit = np.array([3000, 8000, 1000, 10000])

#### Answer- A bar plot, also known as a bar chart or bar graph, is a graphical representation of data that uses rectangular bars or columns to display the values of different categories or groups. Bar plots are commonly used to visualize and compare data across categories, making it easy to understand the relationships and differences between them.

```python

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])


plt.subplot(2, 1, 1)
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.title('Vertical Bar Graph')
plt.show()
```

```python
plt.subplot(2, 1, 2)
plt.barh(company, profit, color='lightgreen')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.title('Horizontal Bar Graph')
plt.show
```

## Q5What is a box plot? Why is it used? Using the following data plot a box plot.
#### box1 = np.random.normal(100, 10, 200)
#### box2 = np.random.normal(90, 20, 200)

#### Answer- A box plot, also known as a box-and-whisker plot, is a graphical representation of data that provides a summary of the distribution, central tendency, and variability of a dataset. It is used to visualize the spread and characteristics of a dataset, especially when dealing with multiple groups or categories

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.boxplot((box1, box2), labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.show()
```
